```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

```python
df = pd.read_csv('D:/AI/data_classification.csv')
df.head()
```

```python
# Split into class 0 & class 1 dataframes
cls0_df = df[df['class'] == 0]
cls1_df = df[df['class'] == 1]
```

```python
# Plot (scatter plot)
ax0 = cls0_df.plot(kind='scatter', x='x', y='y', figsize=(9,6), color='red')
ax1 = cls1_df.plot(kind='scatter', x='x', y='y', figsize=(9,6), color='blue', ax=ax0)
plt.title('Data Classification plot')
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['0', '1'], loc='upper left')
plt.show()
```

```python
df_gpa = pd.read_csv('D:/AI/SAT-GPA.csv')
df_gpa.head()
```

```python
ax = df_gpa.plot(kind='scatter', x='SAT', y='GPA', figsize=(10,7), color='darkblue')
ax.set_title('SAT - GPA Relationship')
ax.set_xlabel('SAT')
ax.set_ylabel('GPA')
x = df_gpa['SAT']
fit = np.polyfit(x, y=df_gpa['GPA'], deg=1)
plt.plot(x, fit[0]*x+fit[1], color='red')
plt.annotate('y={0}x + {1}'.format(fit[0], fit[1]), xy=(1700,2.9))
plt.show()
```

```python
fig = plt.figure()
ax0 = fig.add_subplot(1, 2, 1)
ax1 = fig.add_subplot(1, 2, 2)
df_gpa['SAT'].plot(kind='box', figsize=(16,6), color='green', ax=ax0)
ax0.set_title('SAT Distribution')
ax0.set_ylabel('SAT Score')
df_gpa['GPA'].plot(kind='box', figsize=(16,6), color='green', ax=ax1)
ax1.set_title('GPA Distribution')
ax1.set_ylabel('GPA Score')
plt.show()
```
